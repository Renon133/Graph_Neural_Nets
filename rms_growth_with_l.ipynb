```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
# Parameters
num_steps = 1000  # Length of the random walk
num_simulations = 1000  # Number of random walks to simulate

# Simulate 1D random walks
random_walks_1D = np.cumsum(np.random.choice([-1, 1], size=(num_simulations, num_steps)), axis=1)
displacements_1D = np.mean(random_walks_1D**2, axis=0)
rms_displacement_1D = np.sqrt(displacements_1D)

# Simulate 2D random walks
random_walks_2D = np.cumsum(np.random.choice([-1, 1], size=(num_simulations, num_steps, 2)), axis=1)
displacements_2D = np.mean(np.sum(random_walks_2D**2, axis=2), axis=0)
rms_displacement_2D = np.sqrt(displacements_2D)

# Plotting
plt.figure(figsize=(10, 6))

# 1D RMS displacement
plt.plot(range(1, num_steps + 1), rms_displacement_1D, label="1D RMS Displacement", alpha=0.8)

# 2D RMS displacement
plt.plot(range(1, num_steps + 1), rms_displacement_2D, label="2D RMS Displacement", alpha=0.8)

# Theoretical sqrt(l) for comparison
plt.plot(range(1, num_steps + 1), np.sqrt(range(1, num_steps + 1)), label="Theoretical sqrt(l)", linestyle="--", color="black")

# Labels and legend
plt.xlabel("Number of Steps (l)")
plt.ylabel("Root Mean Square (RMS) Displacement")
plt.title("Growth of RMS Displacement with Length of Random Walk")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Simualte 3D random walks
random_walks_3D = np.cumsum(np.random.choice([-1,1], size= (num_simulations, num_steps, 3)), axis=1)
displacements_3D = np.mean(np.sum(random_walks_3D**2, axis=2), axis=0)
rms_displacement_3D = np.sqrt(displacements_3D)

# Simulate 4d random walks
random_walks_4D = np.cumsum(np.random.choice([-1,1], size= (num_simulations, num_steps, 4)), axis=1)
displacements_4D = np.mean(np.sum(random_walks_4D**2, axis=2), axis=0)
rms_displacement_4D = np.sqrt(displacements_4D)

# Plotting
plt.figure(figsize=(10, 6))
plt.xlabel("Number of Steps (l)")
plt.ylabel("Root Mean Square (RMS) Displacement")
plt.title("Growth of RMS Displacement with Length of Random Walk in 4D")
plt.plot(range(1, num_steps+1), rms_displacement_4D, label="4D RMS Displacement", alpha=0.8)
plt.plot(range(1, num_steps+1), np.sqrt(range(1, num_steps+1)), label="Theoretical sqrt(l)", linestyle="--", color="black")
plt.plot(range(1, num_steps+1), rms_displacement_3D, label="3D RMS Displacement", alpha=0.8)

plt.legend()
plt.grid(True)
plt.show()
```

```python
#Compare RMS in 1D, 2D, and 4D
plt.figure(figsize=(10, 6))
plt.xlabel("Number of Steps (l)")
plt.ylabel("Root Mean Square (RMS) Displacement")
plt.title("Growth of RMS Displacement with Length of Random Walk in 1D, 2D, and 4D")
plt.plot(range(1, num_steps+1), rms_displacement_1D, label="1D RMS Displacement", alpha=0.8)
plt.plot(range(1, num_steps+1), rms_displacement_2D, label="2D RMS Displacement", alpha=0.8)
plt.plot(range(1, num_steps+1), rms_displacement_3D, label="4D RMS Displacement", alpha=0.8)
plt.plot(range(1, num_steps+1), rms_displacement_4D, label="4D RMS Displacement", alpha=0.8)
plt.plot(range(1, num_steps+1), np.sqrt(range(1, num_steps+1)), label="Theoretical sqrt(l)", linestyle="--", color="black")
plt.legend()
plt.grid(True)
plt.savefig("rms_displacement_growth.png")
plt.show()
plt.close()

```
